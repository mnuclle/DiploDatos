4) ***Verificar que los proyectos que tienen nombres duplicados también tienen el hash de nombre duplicado***

```python
from hashlib import md5
```

```python
def hashit(val):
    if isinstance(val, float): 
        return str(val)
    return md5(val.encode('utf-8')).hexdigest()


datasetProjects['hash']=datasetProjects['name'].apply(hashit)
```

```python
datasetProjects[datasetProjects.duplicated('hash')].sort_values('name', ascending=True)
```

```python
datasetProjects[datasetProjects.name=='Pizza'].name.apply(hashit)
```

Definimos DF para verificar y cambiamos el nombre de la columna

```python
df_rep_name_vs_hash = datasetProjects[datasetProjects.duplicated('name')].sort_values('name', ascending=True)
df_rep_name_vs_hash = df_rep_name_vs_hash[['name']].dropna()
df_rep_name_vs_hash.columns = ['rep_name']
df_rep_name_vs_hash.head()
```

Al hacer .info() vemos que tiene 2893 entradas nos va a servir para comparar si se mantienen con los hash repetidos

```python
df_rep_name_vs_hash.info()
```

Armamos el DF que tiene los hash y verificamos repetidos

```python
df_rep_hash = datasetProjects[datasetProjects.duplicated('hash')].sort_values('name', ascending=True)
df_rep_hash = df_rep_hash[['name','hash']].dropna()
df_rep_hash.head()
```

Aplicamos .info() nuevamente para verificar cantidad de valores y vemos que en la columna name se repite la cantidad de 2893 elementos

```python
df_rep_hash.info()
```

Armamos el dataframe con el join por fila de las dos anteriores

```python
df_rep_control = pandas.concat([df_rep_name_vs_hash,df_rep_hash],axis=1,join='inner')
df_rep_control.head()
```

Aplicamos .info() y vemos nuevamente que tenemos 2893 elementos sin null objects lo que nos indica que coinciden por el index ID de los dataframes.

```python
df_rep_control.info()
```

Aplicamos un condicional para verificar que tenemos los mismos nombres y luego contamos para ver si obtenemos el 2893

```python
df_rep_control['control_names'] = (df_rep_control['rep_name'] == df_rep_control['name']).astype(int)
df_rep_control['control_names'].sum()
```

Al obtener el 2893 podemos ver que los hash duplicados corresponden a los mismos nombres del dataframe original que tenía los names duplicados

```python
Q1 = kickstarter_2018.quantile(0.25)
Q3 = kickstarter_2018.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
kickstarter_2018_out = kickstarter_2018[~((kickstarter_2018 < (Q1 - 1.5 * IQR)) |(kickstarter_2018 > (Q3 + 1.5 * IQR))).any(axis=1)]
kickstarter_2018_out.shape
```
