## Random Walk

Aplicamos el algoritmo visto en clase:

- Serie Pospago

```python
serie_inicio_pos = y_pospago.index[-1]+1
h_pos = 6
test=pd.period_range(start=serie_inicio_pos, periods=h_pos, freq='Q')
pred_pos=pd.Series(np.repeat(y_pospago.iloc[-1], h_pos), index=test) 
pred_pos
```

- Análisis Residuos Pospago

```python
resid_pospago=y_pospago-y_pospago.shift(1)
sigma_pos = resid_pospago.std()
round(sigma_pos,3)
```

Se puede ver que no tienen una distribución normal los residuos, graficando el histograma:

```python
resid_pospago.plot.hist(bins=50);
```

Se arma serie de intervalos

```python
np.arange(1,h_pos+1)
```

```python
intv_pos = pd.concat([pred_pos-stats.norm.ppf(0.975)*sigma_pos*np.sqrt(np.arange(1,h_pos+1)),
                   pred_pos+stats.norm.ppf(0.975)*sigma_pos*np.sqrt(np.arange(1,h_pos+1))], axis=1)
intv_pos.round(3)
```

```python
from plotting import fanchart

fan_chart_pos = (intv_pos[0]+intv_pos[1])/2

fig, ax = fanchart(y_pospago['1980':],fan_chart_pos,intv_pos,intv_pos,intv_pos)

plt.show()
```

### Simple Exponential Smoothing

- Serie Pospago

```python
model_ses_pos = sm.tsa.SimpleExpSmoothing(y_pospago).fit()
fitted_ses_pos = model_ses_pos.fittedvalues
model_ses_pos.summary()
```

**MSE POSPAGO**

```python
mse_ses_pos = sm.tools.eval_measures.mse(y_pospago,fitted_ses_pos)
print(' In-sample fit:')
print(' MSE SERIE POSAPGO     {0:.3f}'.format(mse_ses_pos))
```

Cálculo los parámetros:

```python
N_pos = len(y_pospago)
loglik_ses_pos=-(N_pos/2)*(1+np.log(2*np.pi))-(N_pos/2)*np.log(mse_ses_pos)
aic_ses_pos=-2*loglik_ses_pos+2*3
bic_ses_pos=-2*loglik_ses_pos+np.log(N_pos)*3
print(' Log-likelihood    %.3f' % (loglik_ses_pos))
print(' AIC               {0:.3f}'.format(aic_ses_pos))
print(' BIC               {0:.3f}'.format(bic_ses_pos))
```

Se testean los datos suavizados sobre la serie original

```python
fig, ax= plt.subplots();#figsize=(9,6))

y_pospago.plot(color=red, label='Serie Pospago')
fitted_ses_pos.plot(color='black', label='Simple Exponential Smoothing Fit', alpha=0.8)
ax.set_xlabel('')
ax.set_ylabel('Serie Pospago')
ax.set_title('Minutos Pospago Salientes')
ax.set_xticks([], minor=True) 
plt.legend(loc='best')
sns.despine()
plt.show()
```

Como era de esperarse, el comportamiento de la serie suavizada acompaña el de la serie orginal pero los picos y valles son menos pronunciados y están desfazados un período.

- Serie Prepago

```python
model_ses_pre = sm.tsa.SimpleExpSmoothing(y_prepago).fit()
fitted_ses_pre = model_ses_pre.fittedvalues
model_ses_pre.summary()
```

**MSE PREPAGO**

```python
mse_ses_pre = sm.tools.eval_measures.mse(y_prepago,fitted_ses_pre)
print(' In-sample fit:')
print(' MSE SERIE PREPAGO   {0:.3f}'.format(mse_ses_pre))
```

Cálculo de los parámetros:

```python
N_pre = len(y_prepago)
loglik_ses_pre=-(N_pre/2)*(1+np.log(2*np.pi))-(N_pre/2)*np.log(mse_ses_pre)
aic_ses_pre=-2*loglik_ses_pre+2*3
bic_ses_pre=-2*loglik_ses_pre+np.log(N_pre)*3
print(' Log-likelihood    %.3f' % (loglik_ses_pre))
print(' AIC               {0:.3f}'.format(aic_ses_pre))
print(' BIC               {0:.3f}'.format(bic_ses_pre))
```

Se testean los datos suavizados sobre la serie original

```python
fig, ax= plt.subplots();#figsize=(9,6))

y_prepago.plot(color=red, label='Serie Prepago')
fitted_ses_pre.plot(color='black', label='Simple Exponential Smoothing Fit', alpha=0.8)
ax.set_xlabel('')
ax.set_ylabel('Serie Prepago')
ax.set_title('Minutos Prepago Salientes')
ax.set_xticks([], minor=True) 
plt.legend(loc='best')
sns.despine()
plt.show()
```

En este caso, se observa que al principio se da el mismo fenomeno mencionado con la serie pospago, pero los últimos quarters no se refleja bien el acompañamiento del comportamiento de la serie pronosticada con la serie original.

### Validación del Modelo

Se compara el modelo de random walk y simple smoothing

- Serie Pospago

```python
(y_pospago.index.get_loc('2018Q1')-1)/len(y_pospago)
```

Se toma el código visto en clase:

```python
# Real time forecasting - use it as a template - expanding window

validation_pos=y_pospago['2018Q1':].index # the validation period is Q1 2019 onwards
start_pos = y_pospago.index.get_loc('2018Q1') # numerical index corresponding to Q1 2019

pred1 = []
pred2 = []
actual= []
for i in range(start_pos, len(y_pospago)):
    
    actual.append(y_pospago.iloc[i]) # actual value
    
    pred1.append(y_pospago.iloc[i-1]) # random walk forecast
    
    model = sm.tsa.SimpleExpSmoothing(y_pospago.iloc[:i]).fit() 
    pred2.append(model.forecast(1)[0]) # SES forecast

columns=['RW', 'SES', 'Actual']
results_sm = np.vstack([pred1,pred2,actual]).T
results_sm = pd.DataFrame(results_sm, columns=columns, index=validation_pos)
results_sm.tail()
```

En este caso, se observa que el randow walk es mas preciso para esta serie.

```python
from scipy import stats

table_pos = pd.DataFrame(0.0, index=results_sm.columns[:-1], columns=['RMSE','SE'])
for i in range(2):
    table_pos.iloc[i,0] = sm.tools.eval_measures.rmse(results_sm.iloc[:,i], results_sm.iloc[:,-1])
    table_pos.iloc[i,1] = (results_sm.iloc[:,i] - results_sm.iloc[:,-1]).sem()
table_pos.round(3)
```

### Forecast

A continuación se realizan los métodos expanding y rolling forecast para las series.

- Serie Pospago

```python
h_pos=12
w_pos=h_pos*10
data_pos = y_pospago
model = sm.tsa.SimpleExpSmoothing
pred_index = y_pospago.index[-1] + np.array(range(1,h_pos+1))
pred_rolling = pd.Series(0,index=pred_index)
pred_expanding = pd.Series(0,index=pred_index)
orig_exp = None
orig_roll = None
for i in range(0,h_pos):
    current_exp = pd.concat([data_pos, pred_expanding.head(i)])
    expanding = model(current_exp).fit()
    pred_expanding.update(expanding.forecast(1))
    current_roll = pd.concat([data_pos, pred_rolling.head(i)])
    rolling = model(current_roll.tail(w_pos)).fit()
    pred_rolling.update(rolling.forecast(1))
    if i==0:
        orig_fit = expanding.fittedvalues
        orig_roll = rolling.fittedvalues
ax = data_pos[-w_pos:].plot(color='red');#figsize=(9,6))
pd.concat([orig_fit[-w_pos:],pred_expanding]).plot(ax=ax,color='blue');
pd.concat([orig_roll[-w_pos:],pred_rolling]).plot(ax=ax,color='green');
```

- Serie Prepago

```python
h_pre=12
w=h_pre*10
data = y_prepago
model = sm.tsa.SimpleExpSmoothing
pred_index = y_prepago.index[-1] + np.array(range(1,h_pre+1))
pred_rolling = pd.Series(0,index=pred_index)
pred_expanding = pd.Series(0,index=pred_index)
orig_exp = None
orig_roll = None
for i in range(0,h_pre):
    current_exp = pd.concat([data, pred_expanding.head(i)])
    expanding = model(current_exp).fit()
    pred_expanding.update(expanding.forecast(1))
    current_roll = pd.concat([data, pred_rolling.head(i)])
    rolling = model(current_roll.tail(w)).fit()
    pred_rolling.update(rolling.forecast(1))
    if i==0:
        orig_fit = expanding.fittedvalues
        orig_roll = rolling.fittedvalues
ax = data[-w:].plot(color='red');#figsize=(9,6))
pd.concat([orig_fit[-w:],pred_expanding]).plot(ax=ax,color='blue');
pd.concat([orig_roll[-w:],pred_rolling]).plot(ax=ax,color='green');
```
